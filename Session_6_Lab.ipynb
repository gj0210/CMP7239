```python
# Task 3 answer here
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np




df = pd.read_csv("Cybersecurity Intrusion Detection.csv")
print(df.head(10))
```

```python
print("Dataset Info :-")
df.info()
print("n/Descriptive statistics")
display (df.describe())
print("n/Missing values")
display(df.isnull().sum())

plt.figure(figsize=(6,4))
df['attack_detected'].value_counts().plot(kind='bar')
plt.title('Distribution of Attack Detected (0: NO Aattck, 1 Attack)')
plt.xlabel('Attack Detected')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()



```

```python
print("n/Intrusion counts")
print(df['attack_detected'].value_counts())
```
